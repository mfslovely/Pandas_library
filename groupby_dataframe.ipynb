```python
# find number of moives starting with A for each groups


def foo(group):
    print(group)
    return group
```

```python
genres.apply(foo)
```

```python
# find number of moives starting with A for each group

def foo(group):
    return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(foo)
```

```python
# find ranking of each moive in the group according to IMDB score

def rank_moive(group):
    group['geners_rank']=group['IMDB_Rating'].rank(ascending = False)
    return group
```

```python
genres.apply(rank_moive)
```

```python
# find normalized IMDB rating group wise

def normal(group):
    group['non_rating'] = group['IMDB_Rating'].min()/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
    return group
```

```python
genres.apply(normal)
```
