```python
# inner join  or merge - inner join when we can use where in the two table at least one column should be commen
# how - > it tells in pands which type of join you want to perform and 
# on -> it tell which colums you wanna to join 

students.merge(regs,how='inner',on='student_id') 
```

```python
# left join -> it will return matching data from left side 

courses.merge(regs,how='left',on="course_id")
```

```python
# right join -> it will give mathcing data from right side 

temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name': ['Lovely','Nitish','Rahul'],
    'partner':[28,26,17]
})

students = pd.concat([students,temp_df],ignore_index=True)
```

```python
students.tail()
```

```python
students.merge(regs,how='right',on='student_id')
```

```python
# full join - >it will give all matching or common data from left or right join 

students.merge(regs,how='outer',on= 'student_id').tail(10)
```

```python
# find total revenue generated

total = regs.merge(courses,how='inner',on= 'course_id')['price'].sum()
total
```
